```python
from pathlib import Path
from jazzmus.dataset.encoding_convertions import GtParser
from collections import Counter

gt_path = Path("../data/jazzmus/gt")
```

```python
gt_parser = GtParser(split_enc=True, process_harm=True)
```

```python
symbol_counter = Counter()
for file in gt_path.iterdir():
    symbol_counter.update(gt_parser.convert(file))
```

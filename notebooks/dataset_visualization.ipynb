```python
from pathlib import Path
from collections import Counter

dataset_path = Path("../data/ismir_25_jazzmus_cleaned/")
```

```python
print("Number of pieces in the dataset:")
len(list(dataset_path.glob("*.musicxml")))
```

```python
[str(song.stem) for song in dataset_path.glob("*.musicxml") if str(song.stem).startswith("ain_t")]
```

```python
songs = []
for piece_path in dataset_path.glob("*.musicxml"):
    piece_name = piece_path.stem
    if "version" in piece_name:
        piece_name = piece_name.split("_version")[0]
    songs.append(piece_name)

unique_songs = Counter(songs)
print("Number of unique pieces in the dataset:", len(unique_songs))
unique_songs

# song_start = Counter([song[:5] for song in songs])

# song_start

```

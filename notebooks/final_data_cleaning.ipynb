```python
final_dataset_path = Path("../data/jazzmus_dataset/")
final_dataset_path.mkdir(parents=True, exist_ok=True)

for json_file in dataset_path.glob("*.json"):
    # first check for say-it-isn_t-so_version_14.json and rename it to say-it-isn-t-so_version_15.json
    if json_file.stem == "say-it-isn_t-so_version_14":
        new_name = "say-it-isn-t-so_version_15.json"
    else:
        # Determine the new file name: check if its stem is in the dictionary.
        unversioned_name = json_file.stem.split("_version")[0]
        has_version = len(json_file.stem.split("_version")) > 1
        if unversioned_name in pieces:
            new_name = pieces[unversioned_name] + ".json"
            if has_version:
                new_name = new_name.replace(".json", "_version" + json_file.stem.split("_version")[1] + ".json")
        else:
            new_name = json_file.name

    destination = final_dataset_path / new_name
    if destination.exists():
        raise Exception(f"File {destination} already exists. Skipping.")
        
    shutil.copy(json_file, destination)
    print(f"Copied {json_file} to {destination}")
```

```python
from pathlib import Path
import json
import shutil
```

```python
untidy_folder = Path("C:/Users/fosca/Downloads/jazzmus/systems_semantic1835108676906563771/14/")
dataset_folder = Path("../data/jazzmus_dataset/")
# iterate over all json files in the folder if they are not called document.json
for file in untidy_folder.rglob("*.json"):
    print(file)
    if file.stem == "document":
        continue
    # read the content of the json file
    with open(file, "r") as f:
        data = json.load(f)
    # the new name of the json file is taken from the "name" field of the json file
    new_name = data["name"] + ".json"
    # copy the file to the dataset folder with the new name
    shutil.copy(file, dataset_folder / new_name)
    print(f"Copied {file} to {dataset_folder / file.name}")
```
